Vendas por data
1. Agrupar os dados por data e somar as vendas
2. Fazer plot

```python
#1
import seaborn as sns
import matplotlib.pyplot as plt
sales_by_date = sales.groupby('date')['sales'].sum()
```

```python
#2
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.lineplot(x='date', y='sales',data=sales_by_date.reset_index())

plt.title('Vendas por data')
plt.xlabel('Data')
plt.ylabel('Vendas')
plt.xticks(rotation=45)
plt.show()
```
